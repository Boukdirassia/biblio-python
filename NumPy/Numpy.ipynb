```python
import numpy as np
print(np.__version__)
```

```python
# Données linéaires (séries temporelles, caractéristiques d’un modèle ML)
vecteur = np.array([20,30,40,50])
print(vecteur)
print(f"Shape : {vecteur.shape} ")
```

```python
#algèbre linéaire et Machine Learning
matrice_2d = np.array([[20,30,40,50],[20,30,40,50]])
print(matrice_2d)
print(f"Shape : {matrice_2d.shape} ")
```

```python
#vision par ordinateur et Deep Learning (ex: images avec RGB)
matrice_3d = np.array([[[20,30,40],[20,30,40]],[[20,30,40],[30,30,40]]])
print(matrice_3d)
print(f"Shape : {matrice_3d.shape} ")
```

```python
#réseaux de neurones (Deep Learning), analyse volumétrique (IRM, 3D)...
matrice_4d = np.random.rand(2,3,4,5)
print(matrice_4d)
print(f"Shape : {matrice_4d.shape} ")
print(f"Size : {matrice_4d.size} ")
```
