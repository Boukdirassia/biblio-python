```python
data = {'Nom': ["Salma","Assia","Ines","Nada","Laila","Imane"],'Age':[26,22,21,22,20,22],'Poids':[55,57,63,60,45,80]  }
df = pd.DataFrame(data)
print(df)
```

```python
df['Ville']=['Mohammedia','Casablanca','Fes','Meknes','Casablanca','Rabat']
print(df)
```

```python
df['Age'] = df['Age'] + 1
print(df)
```

```python
df_filtre = df[df['Age']>22]
print(df_filtre)
```

```python
df = df.drop(columns='Age')
print(df)
```

```python
df_grouped = df.groupby('Ville')['Poids'].mean()
print(df_grouped)
```

```python
df_sorted = df.sort_values(by='Poids',ascending=True)
print(df_sorted)
```
